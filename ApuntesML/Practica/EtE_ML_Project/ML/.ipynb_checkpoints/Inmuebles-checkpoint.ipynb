# Obtener la información

```python
import os
import tarfile
from six.moves import urllib

# Declaracion de constantes (origen de la info)
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = "datasets/housing"
HOUSING_URL = DOWNLOAD_ROOT + HOUSING_PATH + "/housing.tgz"
```

```python
# Funcion para descargar la informacion
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

fetch_housing_data()
```

```python
import pandas as pd
# Uso de pandas para cargar la informacion
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)

inmuebles = load_housing_data()

# Mostrar por pantalla las primeras lineas (por defecto muestra 5)
inmuebles.head(7)
```

A veces es conveniente separar parte de la información para intentar detectar patrones que te ayuden a decidirte por un algorítmo u otro de ML para implementar en la solución que se esta desarrollando.

Eso es bastante sencillo, con numpy se escoge un 20% (num. arbitrario) y se separa del resto del dataset

```python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

# Separar el dataset entre entrenamiento y comprobacion en un 80/20
train_set, test_set = split_train_test(inmuebles, 0.2)
print(len(train_set), "entrenamiento +", len(test_set), "test")
```

Debido a que cada vez que se ejecute el código el separado puede hacer variar el conjunto de entrenamiento y test, se procede a calcular el hash en base al ID de la instancia y quedarte solo el último byte y poner la instancia en el test si es igual o menor a 51 (que es aproximadamente 20% de 256)

```python
import hashlib

def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]

inmueble_con_id = inmuebles.reset_index() # Se añade una columna indice para utilizarlo como ID
en, te = split_train_test_by_id(inmueble_con_id, 0.2, "index")
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(inmuebles, test_size=0.2, random_state=42)
```

Hasta ahora se han visto métodos aleatorios para la separación de la información para el algorítmo de ML. Debido al volumen de datos no es ningún problema, pero si este fuese bastante menor, correríamos el riesgo de añadir algun tipo de sesgo al separarlo mediante estos métodos

Es importante mantener el ratio de las variables más importantes en la información de entrada para no añadir sesgos inconscientemente. A esto se le conoce como "muestreado estratificado" y la base muestral; subgrupos homogeneos, "estratos".

Suponiendo que el salario medio es una variable importantísima para predecir los valores médios de los inmuebles, podemos aplicar este concepto sobre esta variable. Podemos crear un atributo de la categoría de salarios dividiendo el salario medio entre 1.5 (para limitar la cantidad de categorias) y redondear utilizando "ceil()" y combinar todas las categorias >5 en la categoría 5

```python
inmuebles["income_cat"] = np.ceil(inmuebles["median_income"] / 1.5)
inmuebles["income_cat"].where(inmuebles["income_cat"] < 5, 5.0, inplace=True)
```

Con esto ya esta preparada la información para hacer el muestreado estratificado, el cual podemos conseguir con la clase de scikit-lear StratifiedShuffleSplit

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(inmuebles, inmuebles["income_cat"]):
    strat_train_set = inmuebles.loc[train_index]
    strat_test_set = inmuebles.loc[test_index]
    
inmuebles["income_cat"].value_counts() / len(inmuebles)
```

```python

# Eliminamos la categoría extra que hemos añadido
for set in (strat_train_set, strat_test_set):
    set.drop(["income_cat"], axis=1, inplace=True)
```

### Limpiando la información

La mayoría de algorítmos no soportan que falten características en la información estructurada. Como hemos visto antes, la columna de "total_bedrooms" tiene menos entradas que el resto, asi que para arreglar esto tenemos tres opciones:
- Eliminar esos distritos
- Eliminar el atributo
- Asignar un valor arbitrario (cero, la media, la mediana ...)

```python
inmuebles.dropna(subset=["total_bedrooms"]) # opción 1
inmuebles.drop("total_bedrooms", axis=1) # opción 2
median = inmuebles["total_bedrooms"].median()
inmuebles["total_bedrooms"].fillna(median) # opción 3
```

Scickit-learn tiene una clase implementada que se utiliza para este tipo de tareas "SimpleImputer". Vamos a utilizarla para rellenar con la mediana de los atributos que falten, teniendo en cuenta que esta solo se puede calcular con variables numéricas

```python
from sklearn.impute import SimpleImputer
# Instanciamos el SimpleImputer para rellenar con la mediana
imputer = SimpleImputer(strategy="median")
# Podemos acceder mediante variables de clase a los hyperparametros de esta
print(f"Hyperparametro del imputer: {imputer.strategy}")
# Eliminamos los atributos no numéricos para calcular la mediana
inmuebles_num = inmuebles.drop("ocean_proximity", axis=1)
# Ajustamos el imputer para que calcule la mediana de todas las etiquetas 
imputer.fit(inmuebles_num)
# Podemos acceder mediante variables de clase a los parametros aprendidos de esta
print(f"Parametros aprendidos del imputer: {imputer.statistics_}")
```

```python
# Transformamos el dataset para convertirlo en el de entrenamiento, rellenando los valores faltantes con la media de su etiqueta
train_set = imputer.transform(inmuebles_num)
# Convertimos el resultado, de un array de numpy a un dataset de pandas
housing_tr = pd.DataFrame(train_set, columns=inmuebles_num.columns)
```

A partir de cierto nivel, necesitarás aplicar transformaciones mas complejas o específicas, con lo que tendrás que implementar tus transformers personalizados.

Esto solo conllevará escribir la implementación de los siguientes métodos:
- fit() (return self)
- transform()
- fit_transform()

Se puede conseguir "fit_transform()" sin implementarla si extendemos el transformer personalizado con "TransformerMixin", mientras que con "BaseEstimator" ampliaremos la clase con getters/setters para los el ajuste automático de los hyperparámetros

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    
    def fit(self, X, y=None):
        return self # nothing else to do
    
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
            bedrooms_per_room]
            
        else:
            return np.c_[X, rooms_per_household, population_per_household]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
inmuebles_extra_attribs = attr_adder.transform(inmuebles.values)
```

### Escalado de características

Una de las transformaciones mas importantes a aplicar es el escalado de características, para normalizar los valores numéricos de entrada del algorítmo.

Hay dos maneras comunes de normalizar todos los atributos:
- Min-max: comunente conocido como normalización, ya que el resultado es un valor entre $[0,1]$. Se calcula como $\large x_n = \frac{x_i- x_{min}}{x_{max} + x_{min}}$ (Scikit-Learn tiene MinMaxScaler)
- Estandarización: se resta el valor por la mediana, desplazando la mediana al valor $0$, para despues dividir por la varianza. (Scikit-Learn tiene StandardScaler)

En resumen, min-max calcula un resultado óptimo para ser la entrada de un algorítmo de ML, aun que puede sufrir distorsiones por valores atípicos. Por otro lado, la estandarización no deja un valor en un rango definido pero es más robuso a valores atípicos



### Tuberías de transformaciones

Scikit-Learn provee la clase "Pipeline" para montar secuencias de transformaciones

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
num_pipeline = Pipeline([
('imputer', SimpleImputer(strategy="median")),
('attribs_adder', CombinedAttributesAdder()),
('std_scaler', StandardScaler()),
])
housing_num_tr = num_pipeline.fit_transform(inmuebles_num)
```

La clase "Pipeline" recibe pares de nombre-estimadores, de los cuales todos tienen que ser transformers menos el último. Cuando llamas al método "fit()" del pipeline lo que hace es llamar a todos los "fit_transform()" de los objetos introducidos, pasando los outputs de unos a otros hasta llegar al estimador, en el cual únicamente llama a "fit()"

Una vez rellenados los valores vacios con la mediana, combinados los atributos y escalados los valores numéricos, el último paso será añadir un valor binario a la característica basada en texto, para eso podemos utilizar "FeatureUnion" de Scikit-Learn, el cual recibe una lista de transformers (puede ser una pipeline entera) y ejecuta en paralelo todos los "transform()" para luego concatenar los resultados

```python
from sklearn.base import BaseEstimator, TransformerMixin
# Implementación de la clase para convertir un dataframe de Pandas a array de numpy
class DataFrameSelector(BaseEstimator, TransformerMixin):
    
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
from sklearn.pipeline import FeatureUnion
# Num. atributos y atributo de texto
num_attribs = list(inmuebles_num)
cat_attribs = ["ocean_proximity"]

# Definición de la pipeline completa para transformar valores numéricos
num_pipeline = Pipeline([
('selector', DataFrameSelector(num_attribs)),
('imputer', SimpleImputer(strategy="median")),
('attribs_adder', CombinedAttributesAdder()),
('std_scaler', StandardScaler()),
])

# Definición de la pipeline completa para transformar valores de texto
cat_pipeline = Pipeline([
('selector', DataFrameSelector(cat_attribs)),
('label_binarizer', LabelBinarizer()),
])

# Concatenación de pipelines numérica y de texto
full_pipeline = FeatureUnion(transformer_list=[
("num_pipeline", num_pipeline),
("cat_pipeline", cat_pipeline),
])

#Llamada a la ejecución de la pipeline sobre el dataset
housing_prepared = full_pipeline.fit_transform(inmuebles)
housing_prepared
```
